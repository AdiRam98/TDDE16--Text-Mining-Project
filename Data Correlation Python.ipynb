```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(200)
```

```python
df_numerized.dtypes
```

```python
df['Release Month'] = df['Release date (datetime)'].dt.month
```

```python
df['Season'] = 'Summer'
df['Release Month'].astype(float)
```

```python
df.loc[((df['Release Month']>=1) & (df['Release Month']<=2)), "Season"] = "Winter"
df.loc[(df['Release Month']==12), "Season"] = "Winter"
df.loc[((df['Release Month']>=3) & (df['Release Month']<=5)), "Season"] = "Spring"
df.loc[((df['Release Month']>=9) & (df['Release Month']<=11)), "Season"] = "Autumn"
df["Season"].unique()
```

```python
df['Count']=1
df.info()
```

```python
temp1_df = df.groupby('Release Year')['Season'].apply(lambda x: (x=='Summer').sum()).reset_index(name='count')
temp1_df['Season']='Summer'
temp1_df.head()
```

```python
temp2_df = df.groupby('Release Year')['Season'].apply(lambda x: (x=='Winter').sum()).reset_index(name='count')
temp2_df['Season']='Winter'
temp2_df.head()
```

```python
temp4_df = df.groupby('Release Year')['Season'].apply(lambda x: (x=='Autumn').sum()).reset_index(name='count')
temp4_df['Season']='Autumn'
temp4_df.head()
```

```python
temp3_df = df.groupby('Release Year')['Season'].apply(lambda x: (x=='Spring').sum()).reset_index(name='count')
temp3_df['Season']='Spring'
temp3_df.head()
```

```python
temp_df = pd.concat([temp1_df,temp2_df, temp3_df, temp4_df], ignore_index=True)
temp_df.head()
```

```python
pd.pivot_table(temp_df.reset_index(),
               index='Release Year', columns='Season', values='count'
              ).plot(subplots=True, figsize=(15,12))
plt.show()
```

```python

df_groups = df.groupby('Season')['Unnamed: 0'].count()

df_groups.plot(kind='barh', title='Movies Released By Season',
               xlabel='Season', ylabel='No of Movies', figsize=(10, 6))
for index, value in enumerate(df_groups):
    plt.text(value, index,
             str(value), size=15)

plt.show()
```
